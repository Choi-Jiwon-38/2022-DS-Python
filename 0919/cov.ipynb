y = [2,4,6,8,10] -> 4
-2*-4 + -1*-2 + 0*0 + 1*2 + 2*4 / 8 = 4

선형적인 관계가 있는지 없는지

키 = [183, 180, 165, 160, 190]
몸무게 = [70, 80 65, 40, 90]
서로 관계가 있나? 상관이 있으면 양수값이 나옴

키의 평균 mx
몸무게 평균 my

cov(x,y)= (183-mx)(70-my)+(180-mx)(80-my)+... / 5




x = [2,3,4,5,6]
y = [10,8,6,4,2] -> -4

x = [2,3,4,5,6]
y = [1,1,1,1,1] -> 0

cov(x,y) = -2*0 + -1*0 + ... / 5 = 0

상관계수: cov를 표준편차로 나누어주면 값의 범위가 -1~1로 딱 정해짐 관계가 있다 없다!
1 양의 상관관계가 있다
0 없다
-1 음의 상관관계가 있다



```python
# 상관계수 corr 구하기
f = open("fb_data.txt", "r")

import sys
ids, X, Y = zip(*(tuple(int(x) for x in line.split()) for line in f))

mean_x = sum(X)/len(Y)
mean_y = sum(X)/len(Y)
std_x = (sum((x - mean_x) ** 2 for x in X)/len(X)) ** 0.5
std_y = (sum((y - mean_y) ** 2 for y in Y)/len(Y)) ** 0.5
cov = sum((x - mean_x) * (y - mean_y) for x, y in zip(X, Y)) / len(X)
corr = cov/std_x/std_y

print(corr)
```
