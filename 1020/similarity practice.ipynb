```python
# Pandas, Numpy 패키지 import
import numpy as np
import pandas as pd
```

```python
# Pandas를 이용하여 metadata 읽어오기
meta = pd.read_csv('movies_metadata.csv')
meta
```

```python
# meta 데이터에서 필요한 column만 추출하기
meta = meta[ ['id', 'original_title', 'original_language', 'genres'] ]
meta
```

```python
# 열 이름 변경하기 (rename 사용)
meta = meta.rename(columns = {
    'id': 'movieId',
    'original_title': 'title',
    'original_language': 'language'
})
meta
```

```python
# language가 'en'인 데이터만 추출하기
meta = meta.loc[meta.language == 'en', :]
meta
```

```python
# movieId의 타입을 숫자로 변경하기
meta.movieId = pd.to_numeric(meta.movieId)
meta.movieId
```

```python
# genre 정제하기 - json string을 python set으로 변환하는 함수를 구현
# [ {'id': ~, 'name': ~ }, {...}, {...} ] 꼴의 json string의 형태를 {장르 1, 장르 2, 장르 3}처러 python set의 형식으로 바꾸기 위함

def str_to_set(x):
    # genre_set = set()
    # for item in eval(x):
    #     genre_set.add(item['name'])
    
    genre_set = { n['name'] for n in eval(x)} # 위 코드와 동일함
    
    return genre_set
```

```python
# apply 함수를 이용하여 meta.genres의 모든 값에 str_to_set 적용

meta.genres = meta.genres.apply(str_to_set)
meta.genres
```

```python
keywords = pd.read_csv('keywords.csv')
keywords

# keywords의 keywords도 json string의 형태이므로 python set의 형태로 변환
keywords.keywords = keywords.keywords.apply(str_to_set)
keywords
```

```python
# 향후 join을 위하여 keywords의 id를 movieId로 rename
keywords = keywords.rename(columns = {'id': 'movieId'})
keywords
```

```python
meta = pd.merge(meta, keywords, on = 'movieId', how = 'inner')
meta
```

```python
# Jaccard 유사도를 구하는 함수
def jaccard(a, b):
    if len(a | b) == 0:
        return 0
    return len(a & b) / len(a | b)
```

```python
dk = meta.loc[meta.title == 'The Dark Knight'].iloc[0]
dkr = meta.loc[meta.title == 'The Dark Knight Rises'].iloc[0]

# 두 개를 묶어서 테이블로 변환하는 방법
pd.concat([dk, dkr], axis = 1).T
```

```python
dk_gk = dk.genres | dk.keywords # dk의 장르와 키워드를 합집합으로 만듦
dkr_gk = dkr.genres | dkr.keywords # dkr의 장르와 키워드를 합집합으로 만듦

jaccard(dk_gk, dkr_gk)
```

```python
ratings = pd.read_csv('ratings_small.csv')
ratings
```

```python
ratings =pd.merge(ratings, meta[['movieId', 'title']], on = 'movieId', how = 'inner')
ratings = ratings[ ['userId', 'movieId', 'rating', 'title'] ]
ratings
```

```python
matrix = ratings.pivot_table(index = 'userId', columns = 'title', values = 'rating')
```

```python
dk = matrix['The Dark Knight']
pn = matrix['Prom Night']
```

```python
def pearson_similarity(a, b):
    a = a - a.mean()
    b = b - b.mean()

    denom = (np.sum(a ** 2) * np.sum(b ** 2)) ** 0.5 # 분모

    if denom == 0:
        return 0
    else:
        return np.sum(a * b) / denom
```

```python
pearson_similarity(dk, pn)
```

```python
def find_similar_movies(input_title, matrix, n, alpha):
    input_meta = meta.loc[ meta[ 'title'] == input_title].iloc[ 0]
    input_set = input_meta.genres | input_meta.keywords
    result = []
    for this_title in matrix.columns:
        if this_title == input_title:
            continue

        this_meta = meta.loc[ meta[ 'title'] == this_title].iloc[ 0]
        this_set = this_meta.genres | this_meta.keywords

        pearson = pearson_similarity(matrix[this_title], matrix[input_title])
        jaccard_s = jaccard(this_set, input_set)

        score = alpha * pearson + ( 1-alpha) * jaccard_s
        result.append( (this_title, pearson, jaccard_s, score) )
        
    result.sort(key=lambda r: r[3], reverse=True)
    
    return result[:n]
```

```python
# DarkNightd와 유사한 영화 추천

result = find_similar_movies('The Dark Knight', matrix, 10, 0.3)
pd.DataFrame(result, columns = ['title', 'pearson', 'jaccard', 'score'])
```
